### Ler todas as pipelines Valor

```python
import os
import joblib
import pandas as pd

# Caminho da pasta com os modelos
pasta = r"M:\03-Relatorios\Projecoes\Modelos_valor"

# Listar todos os arquivos .pkl
arquivos = [f for f in os.listdir(pasta) if f.endswith(".pkl")]

resumo = []

for arquivo in arquivos:
    caminho = os.path.join(pasta, arquivo)
    try:
        modelo = joblib.load(caminho)
        
        # Se for pipeline
        if hasattr(modelo, "steps"):
            etapas = [f"{nome}:{est.__class__.__name__}" for nome, est in modelo.steps]
            estimador_final = modelo.steps[-1][1].__class__.__name__
        else:
            etapas = [modelo.__class__.__name__]
            estimador_final = modelo.__class__.__name__
        
        # Monta o dicionário base
        registro = {
            "Arquivo": arquivo,
            "Estimador Final": estimador_final,
        }
        
        # Cria colunas numeradas para cada etapa
        for i, etapa in enumerate(etapas, start=1):
            registro[str(i)] = etapa
        
        resumo.append(registro)
        
    except Exception as e:
        resumo.append({
            "Arquivo": arquivo,
            "Estimador Final": f"Erro: {str(e)[:40]}",
        })

# Criar DataFrame de resumo
df_resumo = pd.DataFrame(resumo)
0
# Exportar para Excel
df_resumo.to_excel(r'M:\03-Relatorios\Projecoes\Estudo projeção\projeçõesvalor.xlsx', index=False)
```
